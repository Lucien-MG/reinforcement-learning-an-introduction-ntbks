# Fibonacci sequence

```python
# Naive implementation of Fibonacci
def fib_naive(n):
    if n <= 1:
        return n
    else:
        return fib_naive(n - 1) + fib_naive(n - 2)
```

```python
%%timeit
fib_naive(10)
```

```python
def fib_map(n, m = {0: 0, 1: 1}):
    if not n in m:
        m[n] = fib_map(n - 1, m) + fib_map(n - 2, m)
    return m[n]
```

```python
%%timeit
fib_map(10)
```

```python
def fib_bottum_up(n):
    if n == 0:
        return 0
    else:
        previousFib, currentFib = 0, 1
        for i in range(n - 1):
            newFib = previousFib + currentFib
            previousFib = currentFib
            currentFib  = newFib
        return currentFib
```

```python
%%timeit
fib_bottum_up(10)
```
